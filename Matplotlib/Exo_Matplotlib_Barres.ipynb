## Diagramme en barres horizontales

```python
# pour un diagramme en barres horizontales, on utilise la méthode Axes.barh(y,width)
# y = position des barres sur l'axe vertical
# width = largeur des barres
```

>- Placer les intitulés de num_cols.
>- Définir l'axe y par 'Sources de notes'.
>- Définir l'axe x par 'Note moyenne'.

```python

notes = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']
bar_positions = range(len(notes))
bar_width = 0.5

film = data['FILM'][0]
notes_moyennes = data[data['FILM'] == film][notes].mean().values

fig, ax = plt.subplots()

ax.barh(bar_positions, notes_moyennes, height=bar_width, color='blue')

ax.set_yticks(bar_positions)
ax.set_yticklabels(notes)

ax.set_ylabel('Sources de notes')
ax.set_xlabel('Note moyenne')
ax.set_title(f'Note moyenne utilisateurs pour le film {film}')

plt.show()
```

En important `recent-grads.csv`, représentez un diagramme en barres donnant le taux de chômage moyen par Major_category


```python
rg = pd.read_csv("https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/recent-grads.csv")
rg.head()
```
