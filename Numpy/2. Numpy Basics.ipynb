# Masque Booléen (appliquer condition sur array)

```python
import numpy as np
a_2 = np.array([[6, 5],
                [1, 3],
                [5, 6],
                [np.nan, 4]])
```

Un masque booléen me permet de filtrer les données d'un ndarray.  
Ici, je filtre les valeurs qui sont supérieurs ou égale à 5.  
Le résultat est une réponse de type booléen pour chaque élément.  
Si la condition est respectée, True, sinon, False.

```python
a_2 >= 5
```

```python
# on peut utiliser any et all avec numpy aussi
print(np.all(a_2 >= 5))
np.any(a_2 >= 5)
```

Je peux utiliser cet ensemble de réponse booléennes comme un "masque".  
Je peux appliquer ce masque au ndarray que je souhaite filtrer.

```python
# Ici je vais ressortir uniquement les valeurs supérieures ou égales à 5
a_2[a_2 >= 5]
```

Ici je veux ressortir uniquement les valeurs qui se sont pas 'nan' (not a number)

**Avec Pandas et Numpy, on exprime la négation avec ~, le AND avec & et le OR avec |**

```python

new_array = a_2[~np.isnan(a_2)]
print(new_array)

# supérieur ou égale à 5 OU 'nan'
new_array = a_2[(np.isnan(a_2)) | (a_2 >= 5)]
print(new_array)

# strictement supérieur à 2 ET strictement inférieur à 5
new_array = a_2[(a_2 > 2) & (a_2 < 5)]
print(new_array)
```

Ainsi, j'ai bien filtrer les valeurs souhaités.

Je peux également faire de la réassignation.  
Ici, je réassigne une nouvel valeur, pour tout les éléments qui respectent ma condition.

```python
a_2[a_2 >= 5] = 30
print(a_2)

# remplacer les valeurs NaN par 0
a_2[np.isnan(a_2)] = 0
print(a_2)
```

## Exo : comparer les consommations annuelles d'alcool par pays et déterminer le pays qui a la plus grande consommation d'alcool

```python
import numpy as np
data = np.genfromtxt("Data/world-alcohol.csv", delimiter = ",", dtype="U75", skip_header =1)
data = data[data[:,4]!='']
print(data.shape)
data
```

```python
# une solution possibe parmi les 100aines qui existent : chacun la sienne donc !
# on va créer un dictionnaire de la forme :
# conso = {pays1 :{annee1 : conso1,
#                  annee2 : conso2,
#                  etc},
#           pays2 :{annee1 : conso1,
#                   annee2 : conso2,
#                   etc},
#            ...}
            
conso = {}

for pays in np.unique(data[:,2]):
    conso[pays] = {}
    for ann in np.unique(data[:,0]):
        filtre = (data[:,2]==pays) & (data[:,0]==ann)
        conso[pays][ann] = data[filtre,4].astype(float).sum()

conso['France']
#conso
```

Pour déterminer le pays avec la plus grande consommation d'alcool, il y a plusieurs mesures possibles :
 - le plus grand consommateur annuel (la plus grande consommation sur une année) ? *réponse le recordman : tel pays a consommé tant en telle année*
 - le plus grand consommateur annuel en moyenne sur toute la période ? *plus grande consommation consommation annuelle moyenne calculée sur 84-89*
 - le plus grand en consommation totale sur la période ?
 - celui qui a été plus grand consommateur annuel le plus de fois sur la période ?
 - système de classement par points avec des points attribués chaque année en fonction de la position
 - etc...

On en implémente quelques unes

```python
### Plus grand en consommation annuelle

# calcul la conso max pour chaque pays
conso_max = {}
for pays in conso:
    conso_max[pays] = max(conso[pays].values())
    
# récupérer la consommation record
conso_record = max(conso_max.values())

# récuéprer le pays et l'année ayant atteint ce record
for pays in conso:
    for ann in conso[pays]:
        if conso_record == conso[pays][ann]:
            pays_record = pays
            annee_record = ann
            break

print(f"{pays_record} a consommé {conso_record} L/hab en {annee_record}")
```

```python
len(conso['France'].values())
```

```python
### Plus grand en consommation myenne sur la période

# calcul la conso moyenne pour chaque pays
conso_moy = {}
for pays in conso:
    conso_moy[pays] = sum(conso[pays].values())/len(conso[pays].values())

# récupérer la plus grande conso moyenne
conso_moy_max = max(conso_moy.values())

# récupérer le pays qui a cette moyenne
for pays in conso_moy:
    if conso_moy[pays] == conso_moy_max:
        pays_max = pays
        break
                
print(f"{pays_max} a la plus grande consommation moyenne sur la période 84-89 avec {conso_moy_max} L/hab consommés en moyenne")
```

```python
### Plus grand en consommation totale sur toute la période

#calcul la conso totale
conso_tot = {}
for pays in conso:
    conso_tot[pays] = sum(conso[pays].values())
    
# récupérer la conso totale max ET le pays associé
pays_max = ''
conso_max = 0

for pays in conso_tot:
    if conso_tot[pays] > conso_max:
        conso_max = conso_tot[pays]
        pays_max = pays

print(f"{pays_max} est le petit champion sur la période 84-89 avec au total {conso_max} L/hab consommés")
```

```python
### autre méthode

# calcul la conso totale pour chaque pays
conso_tot = {}
for pays in conso:
    conso_tot[pays] = sum(conso[pays].values())
    
# récupérer la conso totale max ET le pays associé
conso_max = max(conso_tot.values())

for pays in conso_tot:
    if conso_tot[pays] == conso_max:
        pays_max = pays
        break
                
print(f"{pays_max} est le petit champion sur la période 84-89 avec au total {conso_max} L/hab consommés")
```

```python
### autre méthode
dico = {}
for pays in conso:
    dico[pays] = sum(conso[pays].values())

pays_max = max(dico, key=dico.get)

conso_max =dico[pays_max]
print(f"{pays_max} est le petit champion sur la période 84-89 avec au total {conso_max} L/hab consommés")
```

```python
### Plus grand nombre d'occurences à la 1ère place du podium (sur les 5 années)

# récupérer le pays "vainqueur" et sa consommation pour chaque année
for ann in np.unique(data[:,0]):
    max_ann = 0
    pays_ann = ''

    for pays in conso:
        if conso[pays][ann] > max_ann:
            pays_ann = pays
            max_ann = conso[pays][ann]

    print(ann, pays_ann, max_ann)
```
